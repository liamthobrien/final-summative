## Data Cleaning

```python
# Standardise column names
# Start with country names
# Remove quotes and extra spaces for consistency
df['Area'] = df['Area'].str.replace("'", "").str.strip()

# Clean crop names by removing extra spaces 
df['Item'] = df['Item'].str.strip()

df[['Area', 'Item']].head()
```

```python
# Filter to Wheat data only

# Create a dataframe for wheat production only
wheat_df = df[df['Item'] == 'Wheat']

# Check first few rows
wheat_df.head()
```

```python
# Filter for rows where the element is yield 
wheat_df = wheat_df[wheat_df['Element'] == 'Yield']

wheat_df.head
```

```python
# Handle missing values 
# Check missing values
print("Missing values per column:")
wheat_df.isnull().sum()
```

```python
# Remove flag and note columns (ending in F or N)
# These aren't needed for yield analysis
year_cols_clean = [col for col in wheat_df.columns 
                   if col.startswith('Y') and not col.endswith(('F', 'N'))]

# Keep only Area, Item, Element and clean year columns
keep_cols = ['Area', 'Item', 'Element'] + year_cols_clean
wheat_df_clean = wheat_df[keep_cols]

print(f"Columns after removing flags: {wheat_df_clean.shape[1]}")
```

```python
# Remove rows where ALL year values are missing
# (Keep rows with at least some data)
year_cols_clean = [col for col in wheat_df_clean.columns if col.startswith('Y')]
wheat_df_clean = wheat_df_clean.dropna(subset=year_cols_clean, how='all')

print(f"Rows after removing completely empty records: {len(wheat_df_clean)}")
```

```python
# For remaining missing values, leave them as NaN
# Not all countries grow all crops in all years
print("Final missing value count:")
wheat_df_clean.isnull().sum()
```

```python
# Check final cleaned dataset shape
print(f"Final cleaned dataset shape: {wheat_df_clean.shape}")
```
